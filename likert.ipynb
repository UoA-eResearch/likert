```python
import pandas as pd
import matplotlib.pyplot as plt
import plot_likert
plot_likert.__internal__.BAR_LABEL_FORMAT = "%.0f"
from plot_likert import plot_likert
pd.set_option("display.max_columns", None)
pre = pd.read_csv("data/Data Management Planning (DMP) pre-pilot survey.csv", index_col=0).drop(columns="Indigenous data sovereignty, including for Māori and Pacific peoples")
pre.columns = pd.Series(pre.columns).apply(lambda c: c.strip("\xa02") + " (pre)")
post = pd.read_csv("data/Data Management Planning (DMP) post-pilot survey.csv", encoding_errors="ignore", index_col=0)
post.columns = pd.Series(post.columns).apply(lambda c: c.strip("\xa0") + " (post)")
df = pd.concat([pre, post], axis=1)
df = df[~df.index.isna()]
# This dummy column is used to space bars apart
df["dummy"] = pd.NA
df
```

```python
cols = []
for i in range(len(pre.columns)):
  cols.append(pre.columns[i])
  cols.append(post.columns[i])
  cols.append("dummy")
cols
```

```python
df = df[cols]
df
```

```python
scale = ['No awareness', 'Minimal', 'Fair', 'Good', 'Excellent']
colors = ["#ffffff00", "#AA8602","#BAA141", "#D1D1D1", "#98B9DF", "#335D9F"]
# Bar height
width = .8
ytick_fontsize = 8
```

```python
plt.rcParams['figure.figsize'] = (10,10)
#plt.rcParams['figure.dpi'] = 200

ax = plot_likert(df, scale, colors=colors, plot_percentage=True, bar_labels=True, bar_labels_color="black", title=f"Data Management Planning (DMP) pilot survey", legend=False, width=width)
ax.set_yticklabels([label.get_text().replace("(post)", "").replace("(pre)", "") for label in ax.get_yticklabels()], fontsize=ytick_fontsize)
ax.set_yticks(ax.get_yticks()[2::3])
plt.show()

ax = plot_likert(df.dropna(subset="Working with sensitive data (post)"), scale, colors=colors, plot_percentage=True, bar_labels=True, bar_labels_color="black", title=f"Data Management Planning (DMP) pilot survey", legend=False, width=width)
ax.set_yticklabels([label.get_text().replace("(post)", "").replace("(pre)", "") for label in ax.get_yticklabels()], fontsize=ytick_fontsize)
ax.set_yticks(ax.get_yticks()[2::3])
plt.show()
```

```python
phd = ["DMP03", "DMP09", "DMP10", "DMP16", "DMP17"]
ax = plot_likert(df[df.index.isin(phd)], scale, colors=colors, plot_percentage=True, bar_labels=True, bar_labels_color="black", title=f"Data Management Planning (DMP) pilot survey", legend=False, width=width)
ax.set_yticklabels([label.get_text().replace("(post)", "").replace("(pre)", "") for label in ax.get_yticklabels()], fontsize=ytick_fontsize)
ax.set_yticks(ax.get_yticks()[2::3])
plt.show()
```

```python
handles,labels = ax.get_legend_handles_labels()
fig_legend = plt.figure(figsize=(2,2))
fig_legend.legend(handles, labels, loc='center')
plt.axis("off")
```
