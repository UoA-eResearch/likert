```python
import pandas as pd
import matplotlib.pyplot as plt
from plot_likert import plot_likert
pd.set_option("display.max_columns", None)
pre = pd.read_csv("data/Data Management Planning (DMP) pre-pilot survey.csv", index_col=0).drop(columns="Indigenous data sovereignty, including for Māori and Pacific peoples")
pre.columns = pd.Series(pre.columns).apply(lambda c: c.strip("2") + " (pre)")
post = pd.read_csv("data/Data Management Planning (DMP) post-pilot survey.csv", encoding_errors="ignore", index_col=0)
post.columns = pd.Series(post.columns).apply(lambda c: c + " (post)")
df = pd.concat([pre, post], axis=1)
df = df[~df.index.isna()]
df
```

```python
scale = ['No awareness', 'Minimal', 'Fair', 'Good', 'Excellent']
colors = ["#ffffff00", '#e66101','#fdb863','#f7f7f7','#b2abd2','#5e3c99']
```

```python
plt.rcParams['figure.figsize'] = (10,10)
#plt.rcParams['figure.dpi'] = 100
plot_likert(df, scale, colors=colors, bar_labels=True, bar_labels_color="black", title=f"Data Management Planning (DMP) pilot survey ({len(df)} respondents)", legend=False)
plot_likert(df.dropna(), scale, colors=colors, bar_labels=True, bar_labels_color="black", title=f"Data Management Planning (DMP) pilot survey ({len(df.dropna())} respondents)", legend=False)
```

```python
phd = ["DMP03", "DMP09", "DMP10", "DMP16", "DMP17"]
ax = plot_likert(df[df.index.isin(phd)], scale, colors=colors, bar_labels=True, bar_labels_color="black", title=f"Data Management Planning (DMP) pilot survey ({len(phd)} respondents)", legend=False)
```

```python
handles,labels = ax.get_legend_handles_labels()
fig_legend = plt.figure(figsize=(2,2))
fig_legend.legend(handles, labels, loc='center')
plt.axis("off")
```
