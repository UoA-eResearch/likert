```python
import pandas as pd
import matplotlib.pyplot as plt
from plot_likert import plot_likert
pd.set_option("display.max_columns", None)
pre = pd.read_csv("data/Data Management Planning (DMP) pre-pilot survey.csv", index_col=0).drop(columns="Indigenous data sovereignty, including for MƒÅori and Pacific peoples")
pre.columns = pd.Series(pre.columns).apply(lambda c: c.strip("2"))
pre
```

```python
post = pd.read_csv("data/Data Management Planning (DMP) post-pilot survey.csv", encoding_errors="ignore", index_col=0)
post
```

```python
scale = ['No awareness', 'Minimal', 'Fair', 'Good', 'Excellent']
```

```python
plt.rcParams['figure.figsize'] = (7,7)
#plt.rcParams['figure.dpi'] = 100
plot_likert(pre, scale, bar_labels=True, bar_labels_color="snow", title=f"Data Management Planning (DMP) pre-pilot survey ({len(pre)} respondents)")
pre_complete = pre[pre.index.isin(post.index)]
plot_likert(pre_complete, scale, bar_labels=True, bar_labels_color="snow", title=f"Data Management Planning (DMP) pre-pilot survey ({len(pre_complete)} respondents)")
plot_likert(post, scale, bar_labels=True, bar_labels_color="snow", title=f"Data Management Planning (DMP) post-pilot survey ({len(post)} respondents)")
```

```python
phd = ["DMP03", "DMP09", "DMP10", "DMP16", "DMP17"]
pre_phd = pre[pre.index.isin(phd)]
post_phd = post[post.index.isin(phd)]
plot_likert(pre_phd, scale, bar_labels=True, bar_labels_color="snow", title=f"Data Management Planning (DMP) pre-pilot survey ({len(pre_phd)} respondents)")
plot_likert(post_phd, scale, bar_labels=True, bar_labels_color="snow", title=f"Data Management Planning (DMP) post-pilot survey ({len(post_phd)} respondents)")
```
